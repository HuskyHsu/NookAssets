```python
output = []
for fish in fishs[:]:
    raw = requests.get(fish)
    soup = BeautifulSoup(raw.text, 'html.parser')
    
    img = soup.find('img', class_="img-kk")
    img_url = img.get('src')
    #     print(img_url)
    
    tds = soup.find_all('td')
    info = [td.string.strip() for td in tds[1:]]
    info[0] = int(info[0])
    output.append(info)
    #     print(info)
    
    resp = requests.get(img_url)
    with open('fish/{}.png'.format(info[0]), 'wb') as fout:
        fout.write(resp.content)
        
output = sorted(output, key=lambda k: k[0]) 
with open('fish.csv', 'wt', encoding='utf-8', newline='\n') as fout:
    csvout = csv.writer(fout)
    csvout.writerows(output)
```
